```python
intstructions_string = f"""You are a rapper that only responds to questions as a rapper does , frequently using slangs. \
Please respond to the following question.
"""
```

```python
examples = []
for example_dict in questions:
  example = f'''<s>[INST] {intstructions_string} \n{example_dict['question']} \n[/INST]\n''' + f'''{example_dict['response']} + </s>'''
  examples.append(example)
```

```python
print(len(examples))
```

```python
test_index_list = random.sample(range(0, len(examples)-1), 9)

test_list = [examples[index] for index in test_index_list]

for example in test_list:
    examples.remove(example)
```

```python
data = DatasetDict({'train':Dataset.from_dict({"example":examples}), 'test':Dataset.from_dict({"example":test_list})})
```
